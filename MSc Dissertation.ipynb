```python
from scipy import stats
import numpy as np
import pandas as pd
import collections
import matplotlib.pyplot as plt
import statsmodels.api as sm

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X_features,
                                                     y,
                                                     test_size = 0.2,
                                                     random_state = 42)
```

```python
train_identity = X_train['ID']
X_train = X_train.drop(columns = ['ID'])

test_identity = X_test['ID']
X_test = X_test.drop(columns = ['ID'])
```

```python

X_train[['LIMIT_BAL', 'BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6',
        'PAY_AMT1', 'PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']] = scaler.fit_transform(X_train[['LIMIT_BAL', 'BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6',
        'PAY_AMT1', 'PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']])

X_train.head()
```

```python
df.dtypes
```

```python
import statsmodels.api as sm
logit = sm.Logit(y_train, X_train)
logit_model = logit.fit()
```

```python
logit_model.summary()
```

```python
def get_siginificant_vars (lm):
    # Store the p-values and corresponding column names in a dataframe
    var_p_vals_df = pd.DataFrame(lm.pvalues)
    var_p_vals_df["vars"] = var_p_vals_df.index
    var_p_vals_df.columns = ["pvals", "vars"]
    # Filter the column names where p-values is less than 0.05
    return list (var_p_vals_df[var_p_vals_df.pvals <=0.05]["vars"])


significant_vars = get_siginificant_vars (logit_model)
significant_vars

final_logit = sm.Logit(y_train,
              sm.add_constant(X_train [significant_vars])).fit()

final_logit.summary2()

```

```python
y_predicted_df= pd.DataFrame({"Actual_default_class": y_train,
                         "Probability": final_logit.predict(
                         sm.add_constant( X_train[significant_vars]))})
```

```python
y_predicted_df.sample(15, random_state = 50)
```

```python
y_predicted_df["Predicted_default_class"] = y_predicted_df.Probability.map(
                         lambda x: 1 if x > 0.5 else 0)
y_predicted_df.sample(20, random_state = 42)
```

```python
final_output = pd.concat([train_identity, y_predicted_df], axis = 1).dropna()

print (final_output)
```

```python
final_output.sort_values (by = ["Probability"], inplace=True, ascending= False)
```
